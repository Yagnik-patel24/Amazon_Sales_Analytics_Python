```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px 
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Amazon Sale Report.csv")
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(columns = ['index','currency','ship-country','promotion-ids','fulfilled-by','Unnamed: 22','ship-postal-code'],inplace=True)
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['ship-city'].fillna(method='ffill',inplace=True)
```

```python
df['ship-state'].fillna(method='ffill',inplace=True)
```

```python
df['Amount'].fillna(df['Amount'].mean(),inplace=True)
```

```python
df['Courier_Status'].fillna(df['Courier_Status'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
```

```python
# Amazon brand color palette
amazon_orange = "#FF9900"
amazon_black = "#000000"
amazon_gray = "#232F3E"   # Dark header/nav blue-gray
amazon_light = "#F3F3F3"  # Light gray background

# Set Seaborn theme
sns.set_theme(style="whitegrid")  # Clean white grid
```

```python
order_counts = df["Status"].value_counts(normalize=True).head(10) * 100
print(order_counts)

# Define custom colors for each order status
status_colors = {
    "Shipped - Delivered to Buyer": "#FF9900",   # Orange
    "Cancelled": "#B12704",                      # Red
    "Unshipped": "#232F3E",                      # Dark Gray
    "Shipped": "#FFA41C",                        # Light Orange
    "Returned": "#E47911",                       # Alternate Orange
    "Unknown": "#F3F3F3"                         # Light Gray
}

# Match colors with the order of statuses
color_list = [status_colors.get(status, "#999999") for status in order_counts.index]

# Plot bar chart
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=order_counts.index, y=order_counts.values, palette=color_list)

# Add percentage text labels on top of bars
for bar in ax.patches:
    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        height + 1,
        f'{height:.1f}%',
        ha='center',
        va='bottom',
        fontsize=10,
        color='black',
        fontweight='bold'
    )

# Formatting
plt.xlabel("Order Status", fontsize=12, color="#000000")
plt.ylabel("Percentage (%)", fontsize=12, color="#000000")
plt.title("Order Status Distribution", fontsize=14, fontweight='bold', color="#000000")
plt.xticks(rotation=90, color="#000000")
plt.yticks(color="#000000")
plt.tight_layout()
plt.show()
```

```python
top_sales = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(10)
top_sales_percent = (top_sales / top_sales.sum()) * 100

# Average Quantity by those top 10 states
top_qty = df[df['ship-state'].isin(top_sales.index)].groupby('ship-state')['Qty'].mean()

# Calculate means
sales_mean = top_sales_percent.mean()
qty_mean = top_qty.mean()

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(15, 10))

# Bar chart: Total Sales (%)
sns.barplot(x=top_sales_percent.index, y=top_sales_percent.values, ax=axes[0], color='orange')

axes[0].set_title("Top 10 Ship-States by Total Sales (%)", fontsize=12, fontweight='bold')
axes[0].set_ylabel("Sales (%)")
axes[0].set_xlabel("State")
axes[0].tick_params(axis='x', rotation=45)
axes[0].axhline(sales_mean, color='gray', linestyle='--', linewidth=1)
axes[0].text(len(top_sales_percent)-1, sales_mean + 0.5, f"Mean: {sales_mean:.1f}%", 
             ha='right', fontsize=9, color='gray')

# Line chart: Average Quantity
axes[1].plot(top_qty.index, top_qty.values, marker='o', linestyle='-', color='#146eb4')
axes[1].set_title("Average Quantity by Top 10 Ship-States", fontsize=12, fontweight='bold')
axes[1].set_ylabel("Average Quantity")
axes[1].set_xlabel("State")
axes[1].tick_params(axis='x', rotation=45)
axes[1].axhline(qty_mean, color='gray', linestyle='--', linewidth=1)
axes[1].text(len(top_qty)-1, qty_mean + 0.05, f"Mean: {qty_mean:.2f}", 
             ha='right', fontsize=9, color='gray')
plt.tight_layout()
plt.show()
```

```python
# Group by 'State' and calculate total sales and total quantity
top_states = df.groupby('ship-state').agg({'Amount': 'sum', 'Qty': 'sum'}).sort_values(by='Amount', ascending=False).head(10).reset_index()
top_states.columns = ['State', 'Sales', 'Total Quantity']

# Function to highlight the max sales value
def highlight_max(val):
    color = '#1f77b4' if val == top_states['Sales'].max() else ''
    return f'background-color: {color}; color: white;' if color else ''

# Display styled table
styled_table = top_states.style \
    .bar(subset=['Sales'], color="#FF9900") \
    .bar(subset=['Total Quantity'], color='#146eb4') \
    .applymap(highlight_max, subset=['Sales']) \
    .set_caption("Top 10 States by Sales Value and Quantity") \
    .format({'Sales': '{:,.0f}', 'Total Quantity': '{:,.0f}'})

styled_table
```

```python
# Group by 'Ship-City' and calculate total sales and quantity
top_cities = df.groupby('ship-city').agg({'Amount': 'sum', 'Qty': 'sum'}).sort_values(by='Amount', ascending=False).head(10).reset_index()
top_cities.columns = ['City', 'Sales', 'Total Quantity']

# Function to highlight the max sales value
def highlight_max(val):
    color = '#1f77b4' if val == top_cities['Sales'].max() else ''
    return f'background-color: {color}; color: white;' if color else ''

# Display styled table
styled_table = top_cities.style \
    .bar(subset=['Sales'], color="#FF9900") \
    .bar(subset=['Total Quantity'], color='#146eb4') \
    .applymap(highlight_max, subset=['Sales']) \
    .set_caption("Top 10 Ship-Cities by Sales Value and Quantity") \
    .format({'Sales': '{:,.0f}', 'Total Quantity': '{:,.0f}'})

styled_table
```
